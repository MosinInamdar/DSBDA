```python
planets = sns.load_dataset("planets")
planets
```

```python
planets.describe()
```

```python
planets.isnull().sum()
```

```python
mean_orbital_period = planets['orbital_period'].astype('float').mean()
mean_orbital_period
```

```python
planets['orbital_period'].replace(np.NaN, mean_orbital_period, inplace = True)
planets['orbital_period']
```

```python
mean_mass = planets['mass'].astype('float').mean()
mean_mass
planets['mass'].replace(np.NaN, mean_mass, inplace = True)
planets['mass']
```

```python
mean_distance = planets['distance'].astype('float').mean()
mean_distance
planets['distance'].replace(np.NaN, mean_distance, inplace = True)
planets['distance']
```

```python
planets.isnull().sum()
```
